<center><h3 style="background-color: #FFD54F;color:#222222"><i> PART 3: Exploratory Data Analysis (EDA)</i></h3>

```python
# 1. Top 10 most populated countries in 2022
top10=df.sort_values(by='2022_Population',ascending=False).head(10)
plt.figure(figsize=(10 , 6))
sns.barplot(x='2022_Population' , y='Country/Territory', data=top10, palette='viridis')
plt.title('Top 10 Most Populated Countries (2022)')
plt.xlabel('Population (2022)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# 2. Population by Continent (2022)
continent_pop = df.groupby('Continent')['2022_Population'].sum().sort_values(ascending=False)

plt.Figure(figsize=(8, 5))
continent_pop.plot(kind='bar' , color='teal')
plt.title(' Population by Continent (2022)')
plt.ylabel('Total Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

```python
# 3. Distribution of Growth Rates
plt.Figure(figsize=(8, 5))
sns.histplot(df['Growth_Rate'], bins=20 ,kde=True ,color='orange')
plt.title(' Growth Rate Distribution')
plt.xlabel('Growth Rate (%)')
plt.tight_layout()
plt.show()
```

```python
# 4. Top 10 Densely Populated Countries
top10_density=df.sort_values(by='Density_per_kmÂ²' , ascending=False).head(10)
plt.Figure(figsize=(10, 6))
sns.barplot(x='Density_per_kmÂ²', y='Country/Territory', data=top10_density, palette='magma')
plt.title('Top 10 Densely Populated Countries')
plt.xlabel('Population Density (per kmÂ²)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

<center><h3 style="background-color: #FFD54F;color:#222222"><bold><i>  Part 4 â€“ Feature Engineering</i></h3>

#### Step 1: Create a New Column â€“ Population Growth

```python
# Creating a new column to calculate total population growth from 1970 to 2022
# Subtract 1970 population from 2022 population
df['Population_Growth_(1970_to_2022)'] = df['2022_Population'] - df['1970_Population']
```

#### Step 2: Calculate Average Growth Per Year (Over 52 Years)

```python
# Calculating average yearly population growth
# Total growth / 52 years (from 1970 to 2022)
df['Avg_Growth_Per_Year'] = df['Population_Growth_(1970_to_2022)'] / 52
```

#### Step 3: Calculate Growth Rate from 2000 to 2022

```python
# Calculating population growth rate (in %) from 2000 to 2022
# Formula: ((2022 - 2000 population) / 2000 population) * 100
df['Pop_Growth_Rate_2000_to_2022']=((df['2022_Population'] - df['2000_Population']) / df['2000_Population'] * 100)
```

#### Step 4: Check the Updated DataFrame

```python
# Let's check a few rows to see if new columns were added correctly
df[['Country/Territory', 
    '2022_Population', 
    '1970_Population', 
    '2000_Population', 
    'Population_Growth_(1970_to_2022)', 
    'Avg_Growth_Per_Year', 
    'Pop_Growth_Rate_2000_to_2022']].head(10)
```

<center><h3 style="background-color: #FFD54F;color:#222222"><bold><i>   Part 5 â€“ Correlation Analysis </i></h3>

```python
#Step 1: Calculate Correlation Matrix
# This matrix shows the relationship (correlation) between all numeric features.
# Value close to 1 means strong positive relationship; close to -1 means negative relationship.
correlation_matrix=df.corr(numeric_only=True)
```

```python
#Step 2: Visualize the Correlation Matrix with Heatmap
plt.Figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```
